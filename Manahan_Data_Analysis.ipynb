```python
## Plot Pie chart

job_type_labels = ["Data Scientist", "Data Analyst", "Data Egnineer"]
job_type_values = [2543, 1793, 1379]
job_type_colors = ["coral", "skyblue", "gold"]

plt.pie(job_type_values, labels=job_type_labels, colors=job_type_colors, autopct="%1.1f%%", 
        shadow=True, startangle=140)

plt.savefig("mer_plots/Job_Type_PieChart.png")
```

```python
## display the salary bins

data.Queried_Salary.value_counts()
```

```python
## Plot pie chart

## Show spread of jobs vs salary

job_type_labels = ["<80k","80k-100k", "100k-120k", "120k-140k", "140k-160k",">160k"]
job_type_values = [788, 953, 1394, 1292, 873, 415]
job_type_colors = ["coral", "skyblue", "gold"]

plt.bar(job_type_labels, job_type_values, width=0.8, label="Jobs", color="blue")
plt.xticks(job_type_labels, rotation=30)
plt.ylabel("Number of Jobs")
plt.xlabel("Salary")
plt.legend(loc="upper right")
plt.title("Salary Distribution")
plt.tight_layout()

plt.savefig("mer_plots/Jobs_in_SalaryBin_Bar.png")
```

```python
## No. of Job Postings per Salary bin based on job_type

data.groupby(['Job_Type','Queried_Salary']).Link.count()
```

```python
## Plotting Stacked Bar Chart for salaries for each Job_Type
## Which job_type gets paid the most?

salary_bins = ["<80k","80k-100k", "100k-120k", "120k-140k", "140k-160k",">160k"]
analyst_y = np.array([713, 595, 288, 70, 47, 80])
engineer_y = np.array([27, 125, 396, 452, 275, 104])
scientist_y = np.array([48, 233, 710, 770, 551, 231])
ind = [x for x, _ in enumerate(salary_bins)]

plt.bar(ind, analyst_y, width=0.8, label='Data Analyst', color='skyblue', bottom=engineer_y+scientist_y)
plt.bar(ind, engineer_y, width=0.8, label='Data Engineer', color='gold', bottom=scientist_y)
plt.bar(ind, scientist_y, width=0.8, label='Data Scientist', color='coral')

plt.xticks(ind, salary_bins, rotation=30)
plt.ylabel("Number of Jobs")
plt.xlabel("Salary")
plt.legend(loc="upper right")
plt.title("Salaries vs Job Type")
plt.tight_layout()

plt.savefig("mer_plots/Salaries_vs_JobType.png")
```

```python
## Group the df by Job_Type and Salary


data.groupby(['Job_Type','Queried_Salary']).No_of_Skills.mean().round(1)
```

```python
analyst_skills_y = np.array([3.7, 4.6, 5.9, 6.4, 4.3, 4.4])
engineer_skills_y = np.array([6.9, 9.2, 10.9, 11.3, 11.4, 10.1])
scientist_skills_y = np.array([4.9, 8.2, 8.0, 8.9, 9.2, 8.1])

plt.bar(ind, analyst_skills_y, width=0.8, label='Data Analyst', color='skyblue')
#plt.bar(ind, engineer_skills_y, width=0.8, label='Data Engineer', color='gold')
#plt.bar(ind, scientist_skills_y, width=0.8, label='Data Scientist', color='coral')

plt.xticks(ind, salary_bins, rotation=30)
plt.ylabel("Number of Jobs")
plt.xlabel("Salary")
plt.legend(loc="best")
plt.title("Salaries vs Job Type")
```

```python
## Plot Average Skills Required for Data Analyst Jobs
analyst, = plt.plot(ind, analyst_skills_y, color="skyblue", label="Data Analyst", marker="o", linewidth=2)

## Plot Average Skills Required for Data Engineer Jobs
engineer, = plt.plot(ind, engineer_skills_y, color="gold",label="Data Engineer", marker="X", linewidth=2)

## Plot Average Skills Required for Data Scientist Jobs
scientist, = plt.plot(ind, scientist_skills_y, color="coral", label = "Data Scientist", marker="D", linewidth=2)

# Create a legend for our chart
plt.legend(handles=[analyst, engineer, scientist], loc="lower center")

plt.xticks(ind, salary_bins, rotation=30)
plt.ylabel("Avg No. Skills Required")
plt.xlabel("Salary")
plt.ylim(0,12)
plt.title("Salary vs Avg # of Skills Required")
plt.tight_layout()


plt.savefig("mer_plots/Salaries_vs_AvgSkillReq.png")
```
